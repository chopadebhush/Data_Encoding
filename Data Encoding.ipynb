###  Data Encoding

###  Type of Data Encoding

1. Nominal one hot Encoding
2. Label ordinal Encoding
3. target guided ordinal encoding

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =sns.load_dataset("tips")
df.head()
```

### Nomial One hot Encoding

```python
df.total_bill
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encode_sex =encoder.fit_transform(df[["sex"]]).toarray()
```

```python
encoder.get_feature_names_out()
```

```python
df1 =pd.DataFrame(encode_sex, columns=encoder.get_feature_names_out())
```

```python
final_df =pd.concat([df ,df1],axis=1)
final_df
```

###  label ordinal encoding

```python
from sklearn.preprocessing import LabelEncoder
l_encoder =LabelEncoder()
l_encoder
```

```python
final_df["tip"] =l_encoder.fit_transform(final_df["tip"])
```

```python
final_df
```

####  Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
ord_encoder =OrdinalEncoder()
ord_encoder
```

```python
final_df["time"] =ord_encoder.fit_transform(final_df[["time"]])
```

```python
final_df
```

####  Target Guided ordinal Encoding

```python
mean_price =df.groupby(["time"])["total_bill"].mean().to_dict()
```

```python
df["target_guided_encoded"] =df["time"].map(mean_price)
```
